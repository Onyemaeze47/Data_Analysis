# 1. In the year 2007, provide the country names where the population is between 175000-58000000.

I want to print the first 7 rows of my data frame to get a closer view for analysis

```python
df.head(7)
```

```python
# I am only interested in 2 columns, the 'Country Name' and the '2007' column. I will print this 2 columns;

df1 = df[['Country Name','2007']]
df1
```

```python
# I want to rename the 2007 column so it does not affect my calculation since it is a number. I will call the column Two_Seven.

df1 = df1.rename(columns={'2007':'Two_Seven'}) 
df1.head(7)
```

```python
# There are 266 rows here, i will compare the number of rows in my final answer with this.

# Now lets obtain the dataframe for country name and population in 2007.

new_df1 = df1[(df1.Two_Seven >= 175000) & (df1.Two_Seven <= 58000000)]
new_df1
```

```python
# To print only the 'Country Name', we use the 'drop' function to drop the 'Two_Seven' column

final_df1 = new_df1.drop(['Two_Seven'], axis=1)
final_df1
```

We can see that the number of rows have reduced to 162 in our final answer as compared to 266 when we started.

# 3. For the population in 2003, filter out the population that are in the 3rd quartile (75% and above) and list out the Country Name.

To answer this, we need only the 'Country Name' and the '2003' column. 

```python
# I want to display only the 'Country Name' and '2003' column in the dataset

df3 = df[['Country Name','2003']]
df3
```

```python
# From this, the df3 dataset has 266 rows. I will compare this after my final answer.

# The describe function will provide the third quartile (75% and above) for me, though I would preferably use another method to show it.

df3.describe()
```

```python
# we can see 75% (Third Quartile) is 5.142558e+07  (+07 means times 10 raised to power +7).

# I want to rename the 2003 column so it does not affect my calculation since it is a number. I will call the column Two_Three.

df3 = df3.rename(columns={'2003':'Two_Three'}) 
df3.head(7)
```

```python
# To obtain population in 2003 that are above 75%, we need to obtain values in the third Quartile (Q3)

Q3 = df3.Two_Three.quantile(0.75)
Q3
```

```python
# Notice that the value of Q3 is the same as the 75% value when i did df3.describe().

# Now that i have a base value, Q3 which accounts for 75% of the value in 2003, I will just display the country names that are above this value.

# for convenience i will call the new dataframe dfQ3.
```

```python
dfQ3 = df3[(df3.Two_Three > Q3)]
dfQ3
```

```python
# Now, i would create a new dataframe to drop the 'Two_three' column since it is not needed for my result.
# I will call the new dataframe dfQ3.

new_dfQ3 = dfQ3.drop(['Two_Three'], axis = 1)
new_dfQ3
```
