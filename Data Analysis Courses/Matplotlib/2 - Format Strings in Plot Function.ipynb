```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y, 'g+--')    # you can check out the API documentation for plot function
```

```python
plt.plot(x,y, '--r*')   # the order doesnt matter. You can write it as ('r*--') as well
```

```python
plt.plot(x,y, 'rD')
```

```python
plt.plot(x,y, color='red', marker='D', linestyle='')    # using keywords also yields the same result
```

```python
plt.plot(x,y,color='#008840',marker='D',linestyle='',markersize=12)    #you can also use hexadecimal combination for colors and set marker size
```

```python
plt.plot(x,y, color='#888888', markersize=16, alpha=0.5)  # you can set transparency using alpha (the scale is 0 t0 1)
```
