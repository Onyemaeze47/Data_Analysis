```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ['Home Rent','Food','Phone/Internet Bill','Car','Other Utilities']
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.2f%%')    # you can set autopct to the number of decimal place you want
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.3, autopct='%0.1f%%', shadow=True, explode=[0,0.2,0,0.1,0]) 
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.2,0,0.1,0], startangle=180) 
plt.show()
```
