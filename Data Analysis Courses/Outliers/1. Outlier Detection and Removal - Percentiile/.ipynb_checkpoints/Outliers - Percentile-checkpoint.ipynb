Feature Engineering is a process of extracting useful features from raw data using math, statistics and domain knowledge.

Some of the basic examples are; One Hot Encoding, Outlier Detection and Handling Missing Data

```python
import pandas as pd

df = pd.read_csv('heights.csv')
df
```

```python
max_threshold = df['height'].quantile(0.95)
max_threshold
```

```python
df[df['height'] > max_threshold]
```

```python
min_threshold = df['height'].quantile(0.05)
min_threshold
```

```python
df[df['height'] < min_threshold]
```

```python
df[(df['height'] > min_threshold) & (df['height'] < max_threshold)]
```

```python
df1 = pd.read_csv('C:/Users/user/Downloads/bhps.csv')
df1.head()
```

```python
df1.shape
```

```python
df1.describe()
```

```python
min_threshold, max_threshold = df1.price_per_sqft.quantile([0.001, 0.999])
min_threshold, max_threshold
```

```python
df1[(df1.price_per_sqft < min_threshold) | (df1.price_per_sqft > max_threshold)]
```

```python
df2 = df1[(df1.price_per_sqft > min_threshold) & (df1.price_per_sqft < max_threshold)]
df2.shape
```

```python
df1.shape[0] - df2.shape[0]          # 28 outliers were removed
```

```python
df2.sample(10)
```
